# Credits issued by issuance year (when the registry issued the credits)

Carbon offset registries track offset projects and issue offset credits for each unit of emission reduction or removal that is verified and certified.

```
new_hampshire_baseline = pd.read_csv(r'/work/new_hampshire_GEE_baseline_out_NPP - new_hampshire_GEE_baseline_out_NPP.csv')
new_hampshire_baseline = new_hampshire_baseline.iloc[[10,11,12,13,14,15,16]]
new_hampshire_baseline
```

```
data_preproc = pd.DataFrame({
    'Year': new_hampshire_baseline['year'].to_numpy(), 
    'Baseline Annual NPP': new_hampshire_baseline['annualNPP'].to_numpy(),
    'Observed Annual NPP': observed_newhamp_npp['annualNPP'].to_numpy()})
    
sns.lineplot(x='Year', y='value', hue='variable', 
             data=pd.melt(data_preproc, ['Year']))
plt.title("New Hampshire: Observed NPP & Baseline NPP")
```

```
newhamp_additionality = observed_newhamp_npp['annualNPP'].to_numpy() - new_hampshire_baseline['annualNPP'].to_numpy()

newhamp_additionality
```

```
df_additionality = {'Year':[2014,2015,2016,2017,2018,2019,2020],'Additionality':newhamp_additionality}
df_additionality = pd.DataFrame(df_additionality)
df_additionality
```
